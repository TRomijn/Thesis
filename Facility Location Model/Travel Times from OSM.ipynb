```python
import osrm
import pandas as pd
osrm.__version__
```

```python
osrm.RequestConfig.host = osrm.RequestConfig.host = "router.project-osrm.org"
```

```python
Nepal = pd.read_csv("Data/Nepal Cities Population.csv", usecols=[1,2,3,4], )

# Take only larger cities, because of many null values under 50000 inhabitants. 
# TODO find better dataset with population and coordinates
# Nepal = Nepal [Nepal.Population >= 50000]
```

```python
Nepal = Nepal.dropna()
```

```python
names = list(Nepal.City)
```

```python
latlon = [(i._2, i._3) for i in Nepal.itertuples(False)]
points = [Point(longitude = lat, latitude = lon) for lat, lon in latlon]
```

```python
time_matrix, snapped_coords, a = osrm.table(
    points, ids_origin=names, output='dataframe')
```
